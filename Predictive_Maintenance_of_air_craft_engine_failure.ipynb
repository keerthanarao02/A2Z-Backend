```python
# Creating class labels
# 1 when engine is going to fail within next 30 cycles. 0 otherwise
df['label30']=np.where(df['RUL']<30,1,0)
dfTest['label30']=np.where(dfTest['RUL']<30,1,0)
```

```python
# 70 timesteps (cycles) to produce one sequence
batch70=70
trainIDs=list(df['unit_id'].unique())
inputNP = np.empty((0,18), int)
labelList=[]
```

```python
# Create 3D array from training set for input to LSTM
for id in trainIDs:
    dfTemp=df.loc[df['unit_id']==id]
    start=0
    for i in  range(len(dfTemp)-batch70+1):
        inputNP = np.append(inputNP, dfTemp.iloc[start:start+batch70,1:-2], axis=0)
        labelList.append(dfTemp['label30'].iloc[start+batch70-1])
        start=start+1
trainLSTM = np.reshape(inputNP, (-1,70, 18))
trainLSTM=trainLSTM.astype('float32')
trainLSTM.shape
```

```python
# Array for training labels
trainLabel=np.array(labelList)
trainLabel=np.reshape(trainLabel, (-1,1))
trainLabel.shape
```

```python
# Create 3D array from testing set for evaluation
testIDs=list(dfTest['unit_id'].unique())
inputNP = np.empty((0,18), int)
labelList=[]

for id in testIDs:
    dfTemp=dfTest.loc[dfTest['unit_id']==id]
    start=0
    for i in  range(len(dfTemp)-batch70+1):
        inputNP = np.append(inputNP, dfTemp.iloc[start:start+batch70,1:-2], axis=0)
        labelList.append(dfTemp['label30'].iloc[start+batch70-1])
        start=start+1
testLSTM = np.reshape(inputNP, (-1,70, 18))
testLSTM=testLSTM.astype('float32')
testLSTM.shape
```

```python
# Array for testing labels
testLabel=np.array(labelList)
testLabel=np.reshape(testLabel, (-1,1))
testLabel.shape
```

```python
#Modeling
from tensorflow.keras import regularizers
batch70=70
opt = keras.optimizers.Adam(learning_rate=0.004)
model = Sequential()
model.add(LSTM(input_shape=(batch70, 18), units=100, return_sequences=True, activity_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(LSTM(units=50, return_sequences=False, activity_regularizer=tf.keras.regularizers.l2(0.01)))
model.add(Dense(units=1, activation='sigmoid'))  # Output layer with 1 unit for binary classification
model.compile(loss='binary_crossentropy', optimizer=opt, metrics=[tf.keras.metrics.Recall()])
print(model.summary())
model.save('model.h5')
```

```python
resultTrain = model.fit(
    trainLSTM,
    trainLabel,
    epochs=20,
    batch_size=256,
    validation_split=0.25,
    verbose=1,
    callbacks=[keras.callbacks.EarlyStopping(monitor='val_loss', min_delta=0, patience=20, verbose=0)]
)
```

```python
plt.figure(figsize=(8, 8))
plt.title("Learning curve")
plt.plot(resultTrain.history["loss"], label="training loss")
plt.plot(resultTrain.history["val_loss"], label="validation loss")
plt.plot(resultTrain.history["recall_1"], label="ttraining accuracy")
plt.plot(resultTrain.history["val_recall_1"], label="validation accuracy")
plt.xlabel("Epochs")
plt.ylabel("log_loss")
plt.legend();
```

```python
def get_sensor_data():
    sensor_data = []
    for i in range(1, 24):
        value = float(input(f"Enter sensor s{i} data: "))
        sensor_data.append(value)
    return np.array(sensor_data)

# Main function to predict RUL for user-entered sensor data
def predict_rul_for_user_data():
    while True:
        try:
            # Get the engine unit ID from the user
            engine_id = int(input("Enter engine unit ID: "))

            # Preprocess the input data for the specific engine unit
            input_data_specific = preprocess_data_for_prediction(dfTest, engine_id, batch_size=batch70)

            # Get sensor data from the user
            sensor_data = get_sensor_data()

            # Append the user-entered sensor data to the input data
            input_data_specific[-1] = sensor_data

            # Make predictions using the trained LSTM model
            predictions = model.predict(np.expand_dims(input_data_specific, axis=0))

            # Post-process the predictions to obtain the maximum cycles left (RUL)
            rul_cycles = np.argmax(predictions[:, -1])

            # Display the predicted RUL
            print(f"Predicted maximum cycles left for engine unit {engine_id}: {rul_cycles} cycles")

            # Ask the user if they want to predict RUL for another engine unit
            another_prediction = input("Do you want to predict RUL for another engine unit? (yes/no): ")
            if another_prediction.lower() != 'yes':
                break

        except ValueError:
            print("Invalid input. Please enter valid engine unit ID and sensor data.")

if __name__ == '__main__':
    # Replace 'testIDs' with the list of engine unit IDs in the test set
    testIDs = list(dfTest['unit_id'].unique())

    # Run the prediction loop for user-entered sensor data
    predict_rul_for_user_data()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras

# Load the trained LSTM model (assuming you have already trained it as shown in your code)
model = keras.models.load_model('/content/my_model.h5')  # Replace with the path to your trained model file

# Function to get user input for 18 sensor data values
def get_user_sensor_data():
    user_sensor_data = np.zeros((1, 18), dtype=np.float32)

    for j in range(18):
        user_sensor_data[0, j] = float(input(f"Enter sensor s{j+1} data: "))

    return user_sensor_data

# Get user input for 18 sensor data values
user_sensor_data = get_user_sensor_data()

# Expand the user input to match the model's input shape (70 time steps)
user_sensor_data = np.tile(user_sensor_data, (70, 1))

# Reshape the input to match the model's expected shape
user_sensor_data = np.reshape(user_sensor_data, (1, 70, 18))

# Make predictions based on user sensor data
predictions = model.predict(user_sensor_data)

# Calculate the remaining useful life (RUL) as the time step at which the prediction crosses a threshold
threshold = 0.5  # Adjust this threshold based on your model's performance

predicted_rul = 0
for i in range(70):
    if predictions[0, i] < threshold:
        predicted_rul = i
        break

# Print the predicted RUL
print(f"Predicted Remaining Useful Life (RUL): {predicted_rul} cycles")
```
